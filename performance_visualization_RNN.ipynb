```python
from bokeh.layouts import gridplot, column, row
from bokeh.models import SingleIntervalTicker

def rnn_perf_grid(viz,
                  colors = c,
                  legend_font_size='9pt', 
                     fpr_font_size='9pt', 
                     bestFprOnly=True, 
                     rocs_p_width=800, 
                     rocs_p_height=800,
                     rocs_line_width=3, 
                     rocs_model_appendix = None,
                 spreadsheet_settings={}):

    roc_test_data = [{'fpr':[float(v) for v in viz.loc['test_fpr',model].split(',')[:-1]],
                'tpr':[float(v) for v in viz.loc['test_tpr',model].split(',')[:-1]],
                    'auc':viz.loc['test_auc',model]} for model in sorted(list(viz.columns))]

    test_roc = plot_rocs(roc_test_data, title_lab = 'Test performance', model_appendix = rocs_model_appendix,
               p_width=rocs_p_width, p_height=rocs_p_height, line_width=rocs_line_width,
                    colors = colors, legend_font_size=legend_font_size, fpr_font_size=fpr_font_size,
                   bestFprOnly=bestFprOnly, show_legend=True)
    
    dfs = []
    for model in viz.columns:
        df = pd.DataFrame()
        for par in ['val_acc', 'val_loss', 'acc', 'loss']:
            df[par] = [float(v) for v in viz.loc[par,model].split(',')[:-1]]
        dfs.append((df, model))
    
    #loss_charts = []
    #acc_charts = []

    r_1 = []
    r_2 = []
    for m in range(len(viz.columns)):
        
        ts_loss = timeSeries(dfs[m][0][['loss', 'val_loss']].transpose(), 
                    [c[m], c[m]], 
                    '', 
                    legend_location = 'bottom_right',
                    line_alphas = [0.95, 0.5],
                    line_widths = [4, 2],
                    x_interval_ticker=SingleIntervalTicker(interval=500),
                    x_ticks_type='int', showXGrid=False, plot_w=300, plot_h=300)
        r_1.append(ts_loss)
        #loss_charts.append(ts_loss)
        
        ts_acc = timeSeries(dfs[m][0][['acc', 'val_acc']].transpose(), 
                    [c[m], c[m]], 
                    '', 
                    legend_location = 'bottom_right',
                    line_alphas = [0.95, 0.5],
                    line_widths = [4, 2],
                    x_interval_ticker=SingleIntervalTicker(interval=500),
                    x_ticks_type='int', showXGrid=False, plot_w=300, plot_h=300)
        r_2.append(ts_acc)
        #acc_charts.append(ts_acc)
        #col = column([ts_loss, ts_acc])
        #cols.append(col)
    spr = modelSpreadsheet(**spreadsheet_settings)
    
    gp = gridplot([[row([spr])], [row(test_roc, column(r_1), column(r_2))]])
    return gp
```
