# Buyers and Sellers

This notebook will take the output from step 01 ('01_instrumentsFeatures.ipynb' notebook) and generate separate dataframe of buyers and sellers, in order to perform some data exploration/visualization on the provided portfolio.

## Data import

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import os
```

```python
#Instruments+features dataset
#import data
user = os.environ["USERNAME"]
filename = "190108_instrumentsdf2.pkl"
datafolder = "C:/Users/{}/Dropbox/University/MscDataScience_Birkbeck/thesis_project/data/".format(user)

inst = pd.read_pickle(datafolder+filename)
inst.head()
```

```python
inst.has_impairment1[:10000].sum()
```

```python
#set up buyer dataframe
b_g = inst.groupby("debtor_name_1")
#build an aggregation dictionary

#aggregate by keeping a list
agg_bdict = dict(zip(["customer_id", "customer_name_1", "debtor_id", "uid",
                      "has_impairment1", "has_impairment2", "is_pastdue90", "is_pastdue180",
                      "has_prosecution", "is_open"
                     ],  ["nunique"]*4 + ["sum"]*6  ))

                                                           
#aggregate by first value...
buyer = b_g.agg(agg_bdict)
buyer.head(20)
```

```python
#set up seller dataframes
s_g = inst.groupby("customer_name_1")
#build an aggregation dictionary

#aggregate by keeping a list
agg_sdict = dict(zip(["customer_id", "debtor_name_1", "debtor_id", "uid",
                      "has_impairment1", "has_impairment2", "is_pastdue90", "is_pastdue180",
                      "has_prosecution", "is_open"
                     ],
                   ["nunique"]*4 + ["sum"]*6  ))

                                                           
#aggregate by first value...
seller = s_g.agg(agg_sdict)
seller.head(20)
```

```python
prefix = "190108_"
seller.to_pickle(datafolder + prefix + "seller.pkl")
buyer.to_pickle(datafolder + prefix + "buyer.pkl")

```
