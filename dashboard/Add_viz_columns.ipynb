```python
import pandas as pd
import numpy as np
from os import environ
```

```python
user = environ["USERNAME"]

#Adding columns for visualization info purpose in the repayment patterns graph
datafolder = "C:/Users/{:}/Tradeteq Dropbox/Tradeteq Team/Clients/#GoFactoring/data analysis/".format(user)
filename = "instruments_with_score_simnames.pkl"
tr = pd.read_pickle(datafolder+filename)
```

```python
tr['invoice_date_forVIZ'] = [i[0] for i in tr['invoice_date'].map(str).str.split(' ')]
tr['last_payment_date_forVIZ'] = [i[0] for i in tr['last_payment_date'].map(str).str.split(' ')]
tr['invoice_amount_forVIZ'] = tr['invoice_amount'].map(str)
tr['purchase_amount_forVIZ'] = tr.groupby(tr.index).apply(lambda x:str(np.nansum(x['purchase_amount'])))
tr['due_date_forVIZ'] = [i[0] for i in tr['due_date'].map(str).str.split(' ')]
tr['paid_amount_forVIZ'] = tr.groupby(tr.index).apply(lambda x:str(round(np.nansum([i for i in x['payment_amount']]),2)))
tr['discharge_amount_forVIZ'] = tr.groupby(tr.index).apply(lambda x:str(np.nansum(x['discharge_amount'])))
tr['purchase_amount_open_forVIZ'] = tr.groupby(tr.index).apply(lambda x:str(np.nansum(x['purchase_amount_open'])))
tr['deduction_amount_forVIZ'] = tr.groupby(tr.index).apply(lambda x:str(np.nansum(x['deduction_amount'])))
tr['score_f0_forVIZ']= tr['score_f0'].round(2).map(str)
tr['score_f1_forVIZ']= tr['score_f1'].round(2).map(str)
tr['score_f2_forVIZ']= tr['score_f2'].round(2).map(str)
```

```python
tr.to_pickle(datafolder+filename)
```
